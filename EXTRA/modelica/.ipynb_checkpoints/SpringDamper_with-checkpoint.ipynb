```sysml
package the_sample_model {
    package general_definition {}
    package port_definition {}
    package interface_definition {}
    package system_definition {}
}
```

```sysml
%viz the_sample_model 
```

```sysml
package general_definition{
    //private import attribute_definition::*;
    
    part def mass{
        attribute weight : ISQ::MassUnit;
        attribute position : ISQ::LengthUnit; 
        attribute velocity : ISQ::SpeedUnit;
        }
    part def osilator{
        attribute stifeness;
        attribute C_damp;
    }
    
    part def boundary_condition{
        attribute initial_position : ISQ::LengthUnit;
        attribute initial_velocity: ISQ::SpeedUnit;
    }
    
    part def dynamic_system {
        attribute position : ISQ::LengthUnit; 
        attribute velocity : ISQ::SpeedUnit;
        

    }
    
    item def Force_dynamic {
        item fx {
            attribute value;
            attribute time;
        }
        item fy {
            attribute value;
            attribute time;
        }
        item fz {
            attribute value;
            attribute time;
        }
    }
    item def Velocity_dynamic{
        item vx {
            attribute value;
            attribute time;
        }
        item vy {
            attribute value;
            attribute time;
        }
        item vz {
            attribute value;
            attribute time;
        }
    }
    item def Position_dynamic{
        item px {
            attribute value;
            attribute time;
        }
        item py {
            attribute value;
            attribute time;
        }
        item pz {
            attribute value;
            attribute time;
        }
    }
    item def Acceleration_dynamic{
        item ax {
            attribute value;
            attribute time;
        }
        item ay {
            attribute value;
            attribute time;
        }
        item az {
            attribute value;
            attribute time;
        }
    }
    
}
```

```sysml
%viz --view=DEFAULT --style=LR general_definition
```

```sysml
package port_definition{
    private import general_definition::*;
    
    port def ForcePort{
        out item fx;
        out item fy;
        out item fz;
        in item fx;
        in item fy;
        in item fz;
    }
    port def VelocityPort{
        out item vx;
        out item vy;
        out item vz;
        in item vx;
        in item vy;
        in item vz;
    }
    port def PositionPort{
        out item px;
        out item py;
        out item pz;
        in item px;
        in item py;
        in item pz; 
    }
    port def AccelerationPort{
        out item ax;
        out item ay;
        out item az;
        in item ax;
        in item ay;
        in item az; 
    }
}
```

```sysml
%viz port_definition
```

```sysml
package interface_definition{
    private import general_definition::*;
    private import port_definition::*;
    
    interface def ForceInterface{
        end supplierPort : ForcePort;
        end consumerPort : ForcePort;
    flow of Force_dynamic from supplierPort.fy to consumerPort.fy;
    }
    interface def VelocityInterface{
        end supplierPort : VelocityPort;
        end consumerPort : VelocityPort;
    flow of Velocity_dynamic from supplierPort.vy to consumerPort.vy;
    }
    interface def PositionInterface{
        end supplierPort : PositionPort;
        end consumerPort : PositionPort;
    flow of Position_dynamic from supplierPort.py to consumerPort.py;
    }
    interface def AccelerationInterface{
        end supplierPort : AccelerationPort;
        end consumerPort : AccelerationPort;
    flow of Acceleration_dynamic from supplierPort.ay to consumerPort.ay;
    }   
}
```

```sysml
%viz --view=INTERCONNECTION --style=LR interface_definition_y
```

```sysml
package system_definition{
    private import general_definition::*;
    private import port_definition::*;
    private import interface_definition::*;
    
    part mass_1dof : mass{
        port f : ForcePort; 
        port fDrive : ForcePort;
        port s : PositionPort;
        port v : VelocityPort;
        port a : AccelerationPort;
    }
    part springdamper : osilator{
       port f1 : ForcePort;
       port f2 : ForcePort;
       port s1 : PositionPort;
       port s2 : PositionPort; 
       port v1 : VelocityPort;
       port v2 : VelocityPort; 
   }
    part bc : boundary_condition{
        port s : PositionPort;
        port v : VelocityPort;
        port f : ForcePort;
    }
    

    interface reaction : ForceInterface connect springdamper.f1 to mass_1dof.f ;
    interface reaction : ForceInterface connect bc.f to mass_1dof.fDrive;
    interface position : PositionInterface connect mass_1dof.s to springdamper.s2;
    interface position : PositionInterface connect bc.s to springdamper.s1;
    interface velocity : VelocityInterface connect mass_1dof.v to springdamper.v2;
    interface velocity : VelocityInterface connect bc.v to springdamper.v1;
    


}
```

```sysml
%viz --view=INTERCONNECTION --style=ORTHOLINE --style=TB --style=SHOWIMPORTED  system_definition
```
